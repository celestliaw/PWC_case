Qns: Which Product Category is the best performing?

```python
# Read csv files of Product Category, Products, Order Items and Order Payments
ProductCatName= pd.read_csv("Downloads/data-analytics-case-study-dataset/005_lomo_product_category_name_translation.csv")
Product= pd.read_csv("Downloads/data-analytics-case-study-dataset/004_lomo_products_dataset.csv")
OrderItems= pd.read_csv("Downloads/data-analytics-case-study-dataset/007_lomo_order_items_dataset.csv")
OrderPayments= pd.read_csv("Downloads/data-analytics-case-study-dataset/008_lomo_order_payments_dataset.csv")
```

```python
# Join Product and ProductCatName schemas based on product category
ProductInfo=ProductCatName.merge(Product, left_on= 'product_category_name_portugese', right_on='product_category_name')

ProductInfo.head()

# Group by Product Cateogry name
group_by_frame=ProductInfo.groupby('product_category_name')

group_by_frame.head()
```

```python
# Join OrderItems schema with previous joined datatsets based on common product id
ProductOrder=ProductInfo.merge(OrderItems, left_on='product_id', right_on='product_id')
ProductOrder
```

```python
# Last join with product order schema based on common order id
ProductPayment= ProductOrder.merge(OrderPayments, left_on='order_id', right_on= 'order_id')
ProductPayment
```

```python
# Group by product_name and calculate the total price for each group
# assuming that the installments has taken into account in the payment_value
result = ProductPayment.groupby('product_category_name_english')['payment_value'].sum().reset_index()
result = result.rename(columns={'payment_value': 'total_price'})
# sort values in ascending order based on total price
result.sort_values(by=['total_price'])

```

Qns: As a CMO, should your marketing efforts vary based on states?

```python
#read csv
Customers=pd.read_csv("Downloads/data-analytics-case-study-dataset/001_lomo_customers_dataset.csv")
Sellers=pd.read_csv("Downloads/data-analytics-case-study-dataset/002_lomo_sellers_dataset.csv")
GeoLocation=pd.read_csv("Downloads/data-analytics-case-study-dataset/003_lomo_geolocation_dataset.csv")
ClosedDeals=pd.read_csv("Downloads/data-analytics-case-study-dataset/011_lomo_closed_deals_dataset.csv")
```

```python
# Merge dataframes
#If you want to analyze or visualize the distribution of customers and sellers based on their ZIP code prefixes,
#joining these two columns could help create a comprehensive view of the geographic distribution of your customers and sellers.
merged_df = pd.merge(Customers, Sellers, left_on='customer_zip_code_prefix', right_on='seller_zip_code_prefix', how='inner')
merged_df = pd.merge(merged_df, GeoLocation, left_on='customer_zip_code_prefix', right_on='admin3_pcode', how='inner')
merged_df
```

```python
selected_columns = ['customer_state', 'seller_state', 'shape_length','shape_area', 'shape']
```

```python
# Create a new DataFrame with only the selected columns
new_df = merged_df[selected_columns]
new_df
```

```python
# Concatenate customer and seller states into a single column
all_states = pd.concat([new_df['customer_state'], new_df['seller_state']])

# Create a new DataFrame with state counts
state_counts = all_states.value_counts().reset_index()
state_counts.columns = ['state', 'count']
```

```python
print(state_counts.sort_values(by='count', ascending=False))
```

```python
#counted the occurrences of each state in the merged dataframe and created a bar chart to visualize the distribution of customers and sellers based on their states.
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(state_counts['state'], state_counts['count'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('State Counts Bar Chart')

# Show the plot
plt.show()
```
